```python
# Group by smoker status and calculate mean expenses
expenses_by_smoker = data.groupby('smoker')['expenses'].mean()

# Bar plot
expenses_by_smoker.plot(kind='bar', color=['orange', 'cyan'])
plt.title('Average Expenses of Smokers vs Non-Smokers')
plt.xlabel('Smoker')
plt.ylabel('Average Expenses')
plt.show()
```

```python
# Count smokers by gender
smokers_by_gender = data[data['smoker'] == 'yes'].groupby('sex').size()

# Bar plot
smokers_by_gender.plot(kind='bar', color=['pink', 'purple'])
plt.title('Number of Smokers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Smokers')
plt.show()
```
